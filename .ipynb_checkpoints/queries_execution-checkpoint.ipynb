```python
query = """
WITH current_window AS (SELECT id FROM assessment_windows ORDER BY
academic_year_id DESC, window_type DESC LIMIT 1),
responses AS (
 SELECT tss.grade_level, a.id AS answer_id
 FROM testing_section_students tss
 JOIN users u ON u.id = tss.user_id AND u.role = 7
 JOIN user_answers ua ON ua.testing_section_student_id = tss.id
 JOIN answers a ON a.id = ua.answer_id AND a.question_id = 436
 WHERE tss.subject_area_id = 1 AND tss.academic_year_id = 5
 AND tss.completed_date IS NOT NULL AND tss.grade_level BETWEEN 6 AND 12)
SELECT grade_level,
 CASE answer_id WHEN 1538 THEN 'I believe this' WHEN 1539 THEN 'Struggle to believe'
 WHEN 1540 THEN 'Do not believe' WHEN 1542 THEN 'Did not know' ELSE
 'Other/Unknown' END AS belief_bucket,
 COUNT(*) AS respondents,
 ROUND(100.0*COUNT(*)/SUM(COUNT(*)) OVER (PARTITION BY grade_level),1) AS
 pct_within_grade
FROM responses GROUP BY grade_level, belief_bucket ORDER BY grade_level,
respondents DESC;
"""

df_ex16 = execute_query("ex_16: Student Belief in Real Presence (Grades 6-12)", query)

# Visualization
if df_ex16 is not None and len(df_ex16) > 0:
    pivot_df = df_ex16.pivot(index='grade_level', columns='belief_bucket', values='pct_within_grade').fillna(0)
    pivot_df.plot(kind='bar', figsize=(14, 6), stacked=False)
    plt.title('Student Belief in Real Presence by Grade Level', fontsize=14, fontweight='bold')
    plt.xlabel('Grade Level')
    plt.ylabel('Percentage of Respondents (%)')
    plt.legend(title='Belief', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

# EXAMPLE QUERIES

## ex_01: Average Student Theology Knowledge Score (Diocese)


```python
query = f"""
SELECT AVG(knowledge_percentage) AS avg_score
FROM mv_student_performance
WHERE academic_year_id = 5
 AND subject_area_id = 1
 AND role_category = 'student';
"""

df_ex01 = execute_query("ex_01: Avg Student Theology Knowledge Score", query)
```

## ex_02 & ex_03: Strongest & Weakest Theological Domains


```python
# Strongest Domain
query = f"""
SELECT
 d.name AS strongest_domain,
 AVG(mdp.domain_knowledge_percentage) AS avg_score
FROM mv_domain_performance mdp
JOIN domains d ON d.id = mdp.domain_id
WHERE mdp.academic_year_id = 5
 AND mdp.subject_area_id = 1
 AND mdp.role_category = 'student'
 AND mdp.domain_knowledge_percentage IS NOT NULL
GROUP BY d.id, d.name
ORDER BY avg_score DESC
LIMIT 1;
"""

df_ex02 = execute_query("ex_02: Strongest Theological Domain (Students)", query)

# Weakest Domain
query = f"""
SELECT
 d.name AS weakest_domain,
 AVG(mdp.domain_knowledge_percentage) AS avg_score
FROM mv_domain_performance mdp
JOIN domains d ON d.id = mdp.domain_id
WHERE mdp.academic_year_id = 5
 AND mdp.subject_area_id = 1
 AND mdp.role_category = 'student'
 AND mdp.domain_knowledge_percentage IS NOT NULL
GROUP BY d.id, d.name
ORDER BY avg_score ASC
LIMIT 1;
"""

df_ex03 = execute_query("ex_03: Weakest Theological Domain (Students)", query)
```

## ex_23, ex_24, ex_26: Counts (Students, Teachers, Centers)


```python
# Students
query = """
SELECT COUNT(*) AS total_students
FROM users
INNER JOIN user_testing_centers ON users.id = user_testing_centers.user_id
WHERE user_testing_centers.role = 7
 AND user_testing_centers.deactivate = false
 AND user_testing_centers.testing_center_id IN (
 SELECT id FROM testing_centers WHERE deactivate = false);
"""
df_ex23 = execute_query("ex_23: Total Active Students", query)

# Teachers
query = """
SELECT COUNT(*) AS total_teachers
FROM users
INNER JOIN user_testing_centers ON users.id = user_testing_centers.user_id
WHERE user_testing_centers.role = 5
 AND user_testing_centers.deactivate = false
 AND user_testing_centers.testing_center_id IN (
 SELECT id FROM testing_centers WHERE deactivate = false);
"""
df_ex24 = execute_query("ex_24: Total Active Teachers", query)

# Testing Centers
query = """
SELECT COUNT(*) AS total_active_centers
FROM testing_centers
WHERE deactivate = false;
"""
df_ex26 = execute_query("ex_26: Active Testing Centers", query)
```

## ex_28: Top 5 Schools by Student Knowledge


```python
query = f"""
WITH school_student_scores AS (
 SELECT
 mp.testing_center_id,
 tc.name AS school_name,
 AVG(mp.knowledge_percentage) AS avg_student_score
 FROM mv_student_performance mp
 JOIN testing_centers tc ON tc.id = mp.testing_center_id
 WHERE mp.academic_year_id = 5
 AND mp.subject_area_id = 1
 AND mp.role_category = 'student'
 AND mp.knowledge_percentage IS NOT NULL
 GROUP BY mp.testing_center_id, tc.name
)
SELECT
 ROW_NUMBER() OVER (ORDER BY avg_student_score DESC) AS rank,
 school_name,
 ROUND(avg_student_score::numeric, 2) AS student_knowledge_score,
 testing_center_id
FROM school_student_scores
ORDER BY avg_student_score DESC
LIMIT 5;
"""

df_ex28 = execute_query("ex_28: Top 5 Schools by Student Knowledge", query)

# Visualize
if df_ex28 is not None and len(df_ex28) > 0:
    plt.figure(figsize=(10, 6))
    plt.barh(df_ex28['school_name'], df_ex28['student_knowledge_score'], color='steelblue')
    plt.xlabel('Average Student Knowledge Score')
    plt.title('Top 5 Schools by Student Knowledge', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

## ex_12: Student Knowledge vs Mortal Sins Belief


```python
query = """
SELECT
 CASE a.id WHEN 1520 THEN 'I believe this' WHEN 1522 THEN 'Struggle to believe'
 WHEN 1523 THEN 'Do not believe' WHEN 1524 THEN 'Did not know' ELSE
 'Other/Unknown' END AS confess_mortal_sins_belief,
 AVG((tss.knowledge_score::decimal / NULLIF(tss.knowledge_total,0))*100) AS
 avg_student_knowledge_pct,
 COUNT(DISTINCT tss.id) AS student_test_sessions
FROM testing_section_students tss
JOIN users u ON u.id = tss.user_id AND u.role = 7
JOIN user_answers ua ON ua.testing_section_student_id = tss.id
JOIN answers a ON a.id = ua.answer_id AND a.question_id = 432
WHERE tss.subject_area_id = 1
 AND tss.academic_year_id = 5
 AND tss.completed_date IS NOT NULL
GROUP BY 1
ORDER BY avg_student_knowledge_pct DESC;
"""

df_ex12 = execute_query("ex_12: Student Knowledge vs Mortal Sins Belief", query)
```

```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display, HTML

# --- Step 1: Run query (broaden to include multiple grades) ---
query = """
SELECT
  q.id AS question_id,
  q.body AS question_text,
  a.id AS answer_id,
  a.body AS answer_text,
  tss.grade_level,
  tss.academic_year_id,
  AVG((tss.knowledge_score::decimal / NULLIF(tss.knowledge_total,0)) * 100) AS avg_knowledge_pct,
  COUNT(DISTINCT tss.id) AS student_count
FROM questions q
JOIN answers a ON a.question_id = q.id
LEFT JOIN user_answers ua ON ua.answer_id = a.id
LEFT JOIN testing_section_students tss ON tss.id = ua.testing_section_student_id
LEFT JOIN users u ON u.id = tss.user_id AND u.role = 7
WHERE (tss.subject_area_id = 1 OR tss.subject_area_id IS NULL)
  AND (tss.academic_year_id IN (2, 3, 4, 5) OR tss.academic_year_id IS NULL)
GROUP BY q.id, q.body, a.id, a.body, tss.grade_level, tss.academic_year_id
ORDER BY tss.academic_year_id, tss.grade_level, q.id, a.id;
"""



# --- Step 1: Run the query ---
df = execute_query("Avg Knowledge by Grade, Year, Question & Answer (All Questions)", query)

# --- Step 2: Validate and clean ---
if df is None or len(df) == 0:
    print("No data returned for this query.")
else:
    # Convert and clean
    df['question_id'] = df['question_id'].astype(str)
    df['answer_id'] = df['answer_id'].astype(str)
    df['grade_level'] = df['grade_level'].fillna('Unspecified').astype(str)
    df['academic_year_id'] = df['academic_year_id'].fillna(0).astype(int)
    df['avg_knowledge_pct'] = df['avg_knowledge_pct'].fillna(0)
    df['student_count'] = df['student_count'].fillna(0)

    # --- Map academic years ---
    year_labels = {
        2: '20212022',
        3: '20222023',
        4: '20232024',
        5: '20242025',
        0: 'Unspecified'
    }
    df['academic_year_label'] = df['academic_year_id'].map(year_labels).fillna('Unspecified')

    # --- Dropdowns ---
    dropdown_year = widgets.Dropdown(
        options=sorted([y for y in df['academic_year_label'].unique() if y != 'Unspecified'], reverse=True),
        value='20242025',
        description='Academic Year:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='45%')
    )

    dropdown_grade = widgets.Dropdown(
        options=sorted(df['grade_level'].unique().tolist()),
        value=sorted(df['grade_level'].unique().tolist())[0],
        description='Grade Level:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='30%')
    )

    def build_question_options(grade, year):
        subset = df[(df['grade_level'] == grade) & (df['academic_year_label'] == year)]
        questions = (
            subset[['question_id', 'question_text']]
            .drop_duplicates()
            .sort_values('question_text')
        )

        def truncate(text, n=90):
            return text if len(text) <= n else text[:n-3] + "..."

        return [(truncate(qt), qid) for qid, qt in zip(questions['question_id'], questions['question_text'])]

    question_dropdown = widgets.Dropdown(
        description='Question:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='90%')
    )
    question_dropdown.options = build_question_options(dropdown_grade.value, dropdown_year.value)
    question_dropdown.value = question_dropdown.options[0][1]

    # --- Output chart area ---
    output_chart = widgets.Output()

    # --- Chart update function ---
    def update_chart(*args):
        global fig
        with output_chart:
            output_chart.clear_output(wait=True)
            grade = dropdown_grade.value
            year = dropdown_year.value
            question_id = question_dropdown.value

            filtered = df[
                (df['grade_level'] == grade) &
                (df['academic_year_label'] == year) &
                (df['question_id'] == question_id)
            ]

            if filtered.empty:
                display(HTML("<p style='color:red;'>No data available for this selection.</p>"))
                return

            q_text = filtered['question_text'].iloc[0]
            display(HTML(f"<h4 style='margin-top:10px;'>{q_text}</h4>"))

            filtered = filtered.sort_values('avg_knowledge_pct', ascending=False)
            filtered['status'] = filtered['student_count'].apply(lambda x: 'Answered' if x > 0 else 'No Data')

            fig = px.bar(
                filtered,
                x='avg_knowledge_pct',
                y='answer_text',
                orientation='h',
                color='status',
                color_discrete_map={'Answered': '#4CAF50', 'No Data': '#CCCCCC'},
                text='avg_knowledge_pct',
                hover_data={
                    'answer_id': True,
                    'student_count': True,
                    'avg_knowledge_pct': ':.1f',
                    'answer_text': True
                },
                labels={'avg_knowledge_pct': 'Avg Knowledge %', 'answer_text': 'Answer'},
                height=550
            )

            fig.update_traces(texttemplate='%{text:.1f}%', textposition='outside')
            fig.update_layout(
                title=f"Grade {grade}  {year}: Average Knowledge % by Answer",
                xaxis=dict(title='Average Knowledge %', range=[0, 100]),
                yaxis=dict(title='', automargin=True),
                plot_bgcolor='white',
                margin=dict(l=150, r=30, t=50, b=40),
                legend_title_text='Response Status'
            )

            fig.show()

    # --- Refresh question dropdown on filter change ---
    def on_filter_change(change):
        new_grade = dropdown_grade.value
        new_year = dropdown_year.value
        new_options = build_question_options(new_grade, new_year)
        if new_options:
            question_dropdown.options = new_options
            question_dropdown.value = new_options[0][1]
            update_chart()

    dropdown_year.observe(on_filter_change, names='value')
    dropdown_grade.observe(on_filter_change, names='value')
    question_dropdown.observe(update_chart, names='value')

    # --- Display dashboard ---
    ui = widgets.VBox([
        widgets.HBox([dropdown_year, dropdown_grade]),
        question_dropdown,
        output_chart
    ])

    display(ui)
    update_chart()


```
